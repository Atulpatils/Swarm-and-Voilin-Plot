### Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('gapminder_full.csv')
```

```python
data.head()
```

### Select Features and save them as variable to pass in to seaborn

```python
country=data.country
continent=data.continent
population=data.population
life_exp=data.life_exp
gdp_cap=data.gdp_cap
```

```python
sns.violinplot(x=life_exp)
plt.show()
```

```python
sns.violinplot(x=continent,y=life_exp)
```

### Overlaying Swarmplot Over Violin Plot

```python
plt.figure(figsize=(10,6))
#sns.violinplot(x=continent,y=life_exp)
sns.violinplot(x=continent,y=life_exp)
sns.swarmplot(x=continent,y=life_exp)
plt.title("life expectancy by country")
plt.ylabel('Gapminder life expetancy')
plt.xlabel('Nations')
plt.show()
```
